```python
import cv2
import matplotlib.pyplot as plt
```

```python
from deepface import DeepFace
```

```python
img=cv2.imread('b6.jpg')
```

```python
predictions=DeepFace.analyze(img)
```

```python
predictions
```

```python
a=predictions['region']
x=a['x']
y=a['y']
w=a['w']
h=a['h']
```

```python
color1=(0,0,255)
```

```python
emotion=predictions['dominant_emotion']
yazi2='Emotion:'+emotion
```

```python
gender=predictions['gender']
```

```python
age=predictions['age']
yazi='Age:'+str(age)
```

```python
img=cv2.rectangle(img,(x,y),(x+w,y+h),color1,4)
img=cv2.putText(img,gender,(00,100),cv2.FONT_HERSHEY_SIMPLEX, 3,color1,3)
img=cv2.putText(img,yazi,(00,200),cv2.FONT_HERSHEY_SIMPLEX, 3,color1,3)
img=cv2.putText(img,yazi2,(00,300),cv2.FONT_HERSHEY_SIMPLEX, 3,color1,3)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
